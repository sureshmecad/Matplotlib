# Plotting data from lists using Pyplot API 
# https://www.youtube.com/watch?v=K6eGR3inAm0

```python
# when you pass single parameter, the data is assumed to be Y axis only
# Never forget to add plt.show to the end of plt.object. otherwise you are just creating an object
  & you are not going to see any object.   

views = [534,689,258,401,724,689,350]
plt.plot(views)
plt.show()
```

```python
# Y axis values are same & Index has changed for X axis; Python is using 1 to 7 indexes to plot the data for Y axis.

views = [534,689,258,401,724,689,350]
days = range(1,8)

plt.plot(days, views)
plt.show()
```

# Bar Plot

```python
# The number of views a Youtube channel got in the last 7 days

y_views = [534,689,258,401,724,689,435,534,545,435,423,789,905,561]
x = range(1,15)

plt.bar(x,y_views, label = 'Youtube Views')
plt.xlabel('day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')              # Grid lines generation
plt.title('Daily views for Marketing channels')
plt.show()
```

```python
# The number of views a Youtube channel got in the last 7 days

y_views = [334,589,258,401,724,589,435,534,545,435,423,689,705,561]
f_views = [534,689,558,651,924,789,635,734,745,635,723,789,905,861]
x = range(1,15)

plt.bar(x,y_views, label = 'Youtube Views')
plt.bar(x,f_views, label = 'Facebook Views')
plt.xlabel('day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')              # Grid lines generation
plt.title('Daily views for Marketing channels')
plt.show()
```

```python
# The number of views a Youtube channel got in the last 7 days

y_views = [334,589,258,401,724,589,435,534,545,435,423,689,705,561]
f_views = [534,689,558,651,924,789,635,734,745,635,723,789,905,861]
x = range(1,15)

plt.bar([a-0.25 for a in x], y_views, label = 'Youtube Views')      # (a-0.25 for a in x)
plt.bar([a+0.25 for a in x], f_views, label = 'Facebook Views')     # (a+0.25 for a in x)
plt.xlabel('day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')              
plt.title('Daily views for Marketing channels')
plt.show()
```

```python
# The number of views a Youtube channel got in the last 7 days

y_views = [334,589,258,401,724,589,435,534,545,435,423,689,705,561]
f_views = [534,689,558,651,924,789,635,734,745,635,723,789,905,861]
x = range(1,15)

plt.bar([a-0.25 for a in x], y_views, width = 0.25, label = 'Youtube Views')     # (width = 0.25)
plt.bar([a+0.25 for a in x], f_views, width = 0.25, label = 'Facebook Views')    # (width = 0.25)
plt.xlabel('day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')              
plt.title('Daily views for Marketing channels')
plt.show()
```

```python
# The number of views a Youtube channel got in the last 7 days

y_views = [334,589,258,401,724,589,435,534,545,435,423,689,705,561]
f_views = [534,689,558,651,924,789,635,734,745,635,723,789,905,861]
x = range(1,15)

plt.bar([a-0.25 for a in x], y_views, width = 0.25, label = 'Youtube Views', align ='edge')    # (align ='edge')
plt.bar([a+0.25 for a in x], f_views, width = -0.25, label = 'Facebook Views', align ='edge')  # (align ='edge') & width = -0.25
plt.xlabel('day')
plt.ylabel('Views')
plt.legend(loc = 'upper right')
plt.grid(True, linewidth= 1, linestyle = '-.')              
plt.title('Daily views for Marketing channels')
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0,0].set_title('Upper Left')
axes[0,0].plot([1,2,3,4], [10,20,30,40], color = 'green', linewidth = 3)


x1 = np.linspace (-4, 4)
y1 = x ** 2
axes[0,1].set_title('Upper Right')
axes[0,1].plot(x1,y1)


y2 = x ** 3
axes[1,0].set_title('Lower Left')
axes[1,0].plot(x1,y2)

y3 = x ** 4
axes[1,1].set_title('Upper Right')
axes[1,1].plot(x1,y3)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0,0].set_title('Upper Left')
axes[0,0].plot([1,2,3,4], [10,20,30,40], color = 'green', linewidth = 3)


x1 = np.linspace (-4, 4)
y1 = x ** 2
axes[0,1].set_title('Upper Right')
axes[0,1].plot(x1,y1)


y2 = x ** 3
axes[1,0].set_title('Lower Left')
axes[1,0].plot(x1,y2)

y3 = x ** 4
axes[1,1].set_title('Upper Right')
axes[1,1].plot(x1,y3)

for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
fig, axes=plt.subplots(nrows=2, ncols=2)

x=np.linspace(-4,4)
y1=x
y2=x**2
y3=x**3
y4=x**4

axes[0,0].plot(x, y1)
axes[0,1].plot(x, y2)
axes[1,0].plot(x, y3)
axes[1,1].plot(x, y4)
```
