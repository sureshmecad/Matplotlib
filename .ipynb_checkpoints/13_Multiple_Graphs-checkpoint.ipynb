<h2 style="color:blue" align="left"> Working with Multiple plots </h2>

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
Apl_Price = [93,112,104,144,189]
ms_price = [39,50,57,70,94]
year = [2014,2015,2016,2017,2018]
```

```python
plt.plot(year,Apl_Price, label='Example one')
plt.scatter(year,ms_price, label='Example two',color='g')

plt.legend()

plt.xlabel('year')
plt.ylabel('stock price')

plt.title('info')

plt.show()
```
