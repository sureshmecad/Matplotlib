<h2 style="color:blue" align="left"> Graphs Without Style </h2>

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y,label='skitscat', color='k', s=25, marker="o")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

<h2 style="color:blue" align="left"> Graphs With Style </h2>

```python
from matplotlib import style
style.use('ggplot')

x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y,label='skitscat', color='k', s=25, marker="o")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```
