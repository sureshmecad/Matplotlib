```python
# from matplotlib import pyplot as plt

import pandas as pd
```

```python
data = pd.read_csv("C:/Users/deepusuresh/Documents/Data Science/01. Python/4. Matplotlib/cgpa.csv")
data.head()
```

```python
df = data.head(50)                             # Dataset have total rows of 50,000 & graph comes as very clumpsy

plt.plot(df["rollno"], df["cgpa"], 'r')

plt.xlabel('Roll No')
plt.ylabel('CGPA')
plt.title('Roll No Vs CGPA')                   # since "%matplotlib inline" called, no need to write plt.show()
```

```python
df = data.head(50)                                      # Dataset have total rows of 50,000 & graph comes as very clumpsy

plt.plot(df["rollno"], df["cgpa"], color='r', label="Line Graph")
plt.scatter(df["rollno"], df["cgpa"], color='b', label="Scatter")

plt.xlabel('Roll No', color="green")
plt.ylabel('CGPA', color="blue")
plt.title(' CGPA Vs Roll No', color='green')            # since "%matplotlib inline" called, no need to write plt.show()

plt.legend()
```
