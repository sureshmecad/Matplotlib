<h2 style="color:blue" align="left"> 1. Pylab </h2>

```python
from matplotlib import pylab
import numpy as np
```

```python
x = np.linspace(0,10,25)
y = x * x + 2
```

```python
pylab.plot(x,y,'r')                  # r stands for red
```

<h3 style="color:green" align="left"> Drawing a Subgraph </h3>

```python
pylab.subplot(1,2,1)               # The contents of the brackets represents (rows, columns, indexes)
pylab.plot(x,y,'r--')              # The third parameter here determines color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x,'g*-')
```

<h2 style="color:blue" align="left"> 2. Pyplot </h2>

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5,0.1,0.8,0.8])       # control the left, right, width, height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)         # submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

#### we can also draw a picture or graph, inside another graph

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])      # big axes
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])      # small canvas
axis1.plot(x,y,'r')
axis2.plot(y,x,'g')
```

```python
fig = plt.figure(figsize=(16,9), dpi=300)
fig.add_subplot()
plt.plot(x,y,'r')
```

```python
fig, axes = plt.subplots(figsize=(16,9), dpi=50)
axes.plot(x,y,'r')
```

```python
ax.legend(["label1","label2"])

fig, axes = plt.subplots()

axes.set_xlabel("x-label")
axes.set_ylabel("y-label")

axes.set_title("title of graph")

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(["y=x**2", "y=x**3"], loc=2)       # loc=1  --->  Top Right corner
                                               # loc=2  --->  Top Left corner
                                               # loc=3  --->  Bottom Left corner
                                               # loc=4  --->  Bottom Right corner
```

#### Matplotlib can set other properties such as line color, transperancy and more

```python
fig, axes = plt.subplots(dpi=150)
axes.plot(x, x+1)                      # generating default colors
axes.plot(x, x+2)
axes.plot(x, x+3)
```

```python
fig, axes = plt.subplots(dpi=150)
axes.plot(x, x+1, color='red')               # Defined colors
axes.plot(x, x+2, color='blue')
axes.plot(x, x+3, color='green')
```

```python
# draw contour image

import matplotlib
import numpy as np
import matplotlib.cm as cm        # cm stands for color maps
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
x,y = np.meshgrid(x,y)
z1 = np.exp(-x**2 - y**2)
z2 = np.exp(-(x-1)**2 - (y-1)**2)
z = (z1-z2) * 2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(x, y, z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('contour map')
```

```python
# draw 3d surface image

from mpl_toolkits.mplot3d.axes3d import axes3d

fig = plt.figure(figsize(14,6))

# specify the 3d graphics to draw, with projection='3d'

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(x, y, z, rstride=4, cstride=4, linewidth=0)
```
