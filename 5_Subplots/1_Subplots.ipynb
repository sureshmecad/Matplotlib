```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
y=x*x
```

```python
plt.subplot(2,2,1)           # 2,2,1  --->  rows:2, columns:2, position:1
plt.plot(x,y,'r--')

plt.subplot(2,2,2)           # 2,2,2  --->  rows:2, columns:2, position:2
plt.plot(x,y,'g*--')

plt.subplot(2,2,3)           # 2,2,3  --->  rows:2, columns:2, position:3
plt.plot(x,y,'bo')

plt.subplot(2,2,4)           # 2,2,4  --->  rows:2, columns:2, position:4
plt.plot(x,y,'go')
```

```python
Apl_Price = [93,112,104,144,189]
ms_price = [39,50,57,70,94]
year = [2014,2015,2016,2017,2018]
```

```python
fig_1 = plt.figure(1, figsize=(20,4.8))

chart_1 = fig_1.add_subplot(121)           
chart_2 = fig_1.add_subplot(122)           

chart_1.plot(year,Apl_Price)
chart_2.scatter(year,ms_price)

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(20,9.6))

chart_1 = fig_1.add_subplot(224)  # 224 ---> rows:2, columns:2, position:4
chart_2 = fig_1.add_subplot(221)  # 221 ---> rows:2, columns:2, position:1

chart_1.plot(year,Apl_Price)
chart_2.scatter(year,ms_price)

plt.show()
```

```python
fig_2, axes = plt.subplots(3,2, figsize=(20,10))
```

```python
fig_2, axes = plt.subplots(2,2, figsize=(20,5))
axes[0,1].scatter(year,ms_price)
axes[1,0].plot(year,Apl_Price)
plt.show()
```

```python
# Compute the a and b coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show()
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```
